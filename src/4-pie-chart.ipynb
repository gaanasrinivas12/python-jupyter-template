### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

try:
    # Data
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must have the same length.")
    
    if sum(sizes) == 0:
        raise ValueError("The sum of sizes must be greater than zero.")

    # Plot
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
            colors=['blue', 'red', 'green', 'orange'], wedgeprops={'edgecolor': 'black'})

    # Title
    plt.title('Programming Language Popularity')

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
